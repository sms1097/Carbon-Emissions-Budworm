# Data Cleaning

This notebook is splitting the features into an early version for predictions

```python
import pandas as pd
```

## Load Data

```python
LCA_SCENARIO = 'NetC'
```

```python
data = pd.read_csv('SBW_NetC_Outcomes_May42020.csv')
data = data[data['LCA_Scenario'] == 'NetC']
data.head()
```

```python
data = data.drop(['Time_Scenario', 'LCA_Scenario'], axis=1)
```

```python
data.head()
```

## Create New Dataset
This is where we create the data set that is the expansion of these columns

```python
clean_df = pd.DataFrame(columns=['TimeStep', 'Risk_Cat', 'Stand_ID', 'Salvage', 'Management', 'Result'])

## Iterate through columns and get value
for _, curr_row in data.iterrows():
    row = pd.DataFrame()
    for feature, val in curr_row.iteritems():
        if feature == "TimeStep":
            row['TimeStep'] = [val]
        elif feature == "Risk_Cat":
            row['Risk_Cat'] = [val]
        elif feature == 'Stand_ID':
            row['Stand_ID'] = [val]
        elif feature == 'Heavy_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['Heavy']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'NoMgmt_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['NoMgmt']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Moderate_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['Moderate']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Comm-Ind_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['Comm-Ind']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'HighGrade_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['HighGrade']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Light_Salvage':
            temp = row.copy()
            temp['Salvage'] = [True]
            temp['Management'] = ['Light']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Light_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['Light']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Comm-Ind_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['Comm-Ind']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'NoMgmt_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['NoMgmt']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'HighGrade_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['HighGrade']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Heavy_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['Heavy']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Moderate_NoSalvage':
            temp = row.copy()
            temp['Salvage'] = [False]
            temp['Management'] = ['Moderate']
            temp['Result'] = [val]
            clean_df = clean_df.append(temp)
        elif feature == 'Benchmark_StandC':
            continue
        else:
            print(feature)
            break
```

```python
clean_df.to_csv('Expanded_Features.csv')
```
